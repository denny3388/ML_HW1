```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sklearn
from sklearn.model_selection import train_test_split # for splitting the data into train and test samples
from sklearn.metrics import classification_report # for model evaluation metrics
from sklearn.preprocessing import OrdinalEncoder # for encoding categorical features from strings to number arrays
from sklearn.naive_bayes import CategoricalNB
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import KFold
```

```python
""" Data input """
# Mushroom
name_list1 = ['edible','cap-shape','cap-surface','cap-color','bruises','odor','gill-attachment','gill-spacing','gill-size','gill-color','stalk-shape','stalk-root','stalk-surface-above-ring','stalk-surface-below-ring','stalk-color-above-ring','stalk-color-below-ring','veil-type','veil-color','ring-number','ring-type','spore-print-color','population','habitat']
data1 = pd.read_csv('agaricus-lepiota.data', sep=",", names=name_list1)
feature_list1 = name_list1
feature_list1.remove('edible')
attribute_list1 = [['b','c','x','f','k','s'],['f','g','y','s'],['n','b','c','g','r','p','u','e','w','y'],['t','f'],['a','l','c','y','f','m','n','p','s'],['a','d','f','n'],['c','w','d'],['b','n'],['k','n','b','h','g','r','o','p','u','e','w','y'],['e','t'],['b','c','u','e','z','r'],['f','y','k','s'],['f','y','k','s'],['n','b','c','g','o','p','e','w','y'],['n','b','c','g','o','p','e','w','y'],['p','u'],['n','o','w','y'],['n','o','t'],['c','e','f','l','n','p','s','z'],['k','n','b','h','r','o','u','w','y'],['a','c','n','s','v','y'],['g','l','m','p','u','w','d']]
cate_num_list1 = [6,4,10,2,9,4,3,2,12,2,4,4,9,9,2,4,3,8,9,6,7]

# Iris
name_list2 = ['sepal-length','sepal-width','petal-length','petal-width','class']
data2 = pd.read_csv('iris.data', sep=",", names=name_list2)
data2_group = data2.groupby('class')
feature_list2 = name_list2
feature_list2.remove('class')
```

```python
""" Data preprocessing """

# Drop feature w/ missing value 
data1 = data1.drop(columns='stalk-root')
feature_list1.remove('stalk-root')

# Transform data format
data1['edible_num']=data1['edible'].apply(lambda x: 1 if x=='e' else 0)
data2['class_num']=data2['class'].apply(lambda x: 1 if x=='Iris-setosa' else 
                                       0 if x=='Iris-versicolor' else -1)
# shuffle
data1.sample(frac=1)
data2.sample(frac=1)
```

```python
# Function that handles sample splitting, model fitting and report printing 
def holdout(X, y, typ):
    
    """ Holdout validation """
    # Create training and testing samples
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)

    # Fit the model
    model = typ
    holdout_clf = model.fit(X_train, y_train)
    
    # class 'e' confusion matrix
    TP, FP, FN, TN = 0, 0, 0, 0
    pred = holdout_clf.predict(X_test)
    for i in range(len(X_test)):
        if pred[i] == 'e' and y_test[i] == 'e':
            TP += 1
        elif pred[i] == 'e' and y_test[i] == 'p':
            FP += 1
        elif pred[i] == 'p' and y_test[i] == 'e':
            FN += 1
        elif pred[i] == 'p' and y_test[i] == 'p':
            TN += 1
    confusion_mat = [[TP, FP], [FN, TN]]
    acc = (TP+TN)/len(X_test)
    recall = TP/(TP+FN)
    precision = TP/(TP+FP)
    
    # Use score method to get accuracy of the model
    print('--------------------------------------------------------')
    holdout_score = model.score(X_test, y_test)
    print('Accuracy Score(holdout): ', holdout_score)
    print('--------------------------------------------------------')
    
    # Return relevant data for chart plotting
    return holdout_clf, holdout_score, confusion_mat, acc, recall, precision
```

```python
# Select data for modeling
X=data1[feature_list1]
y=data1['edible'].values

enc = OrdinalEncoder()
X = enc.fit_transform(X)

# Fit the model and print the result
m_holdout_clf, m_holdout_score, m, acc, recall, precision = holdout(X, y, CategoricalNB(alpha=1e-10, min_categories=cate_num_list1)) # without laplace
print(f"matrix = {m}, accuracy = {acc}, recall = {recall}, precision = {precision}")
```

```python
# Function that handles sample splitting, model fitting and report printing 
def mfunc(X, y, typ):
    
    """ Holdout validation """
    # Create training and testing samples
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)

    # Fit the model
    model = typ
    holdout_clf = model.fit(X_train, y_train)
     
    # Use score method to get accuracy of the model
    print('--------------------------------------------------------')
    holdout_score = model.score(X_test, y_test)
    print('Accuracy Score(holdout): ', holdout_score)
    print('--------------------------------------------------------')
    
    """ K-fold """
    K = 3
    kf_clfs = []
    score_sum = 0
    
    kf = KFold(n_splits=K)
    for train_index, test_index in kf.split(X,y):
        X_train, X_test = X[train_index], X[test_index]
        y_train, y_test = y[train_index], y[test_index]

        # Fit the model
        model = typ
        kf_clfs.append(model.fit(X_train, y_train))
        
        score_sum = score_sum + model.score(X_test, y_test)
        
    # Use score method to get accuracy of the model
    print('--------------------------------------------------------')
    kf_score = score_sum / K
    print('Accuracy Score(k-fold): ', kf_score)
    print('--------------------------------------------------------')
    
    # Return relevant data for chart plotting
    return holdout_clf, holdout_score, kf_clfs, kf_score
```

```python
# Select data for modeling
X=data1[feature_list1]
y=data1['edible'].values

enc = OrdinalEncoder()
X = enc.fit_transform(X)

# Fit the model and print the result
m_holdout_clf, m_holdout_score, m_kf_clf, m_kf_score = mfunc(X, y, CategoricalNB(alpha=1e-10, min_categories=cate_num_list1)) # without laplace
m_holdout_laplace_clf, m_holdout_laplace_score, m_kf_laplace_clf, m_kf_laplace_score = mfunc(X, y, CategoricalNB(alpha=1, min_categories=cate_num_list1)) # with laplace
```

```python
# Function that handles sample splitting, model fitting and report printing 
def holdout(X, y, typ):
    
    """ Holdout validation """
    # Create training and testing samples
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)

    # Fit the model
    model = typ
    holdout_clf = model.fit(X_train, y_train)
    
    # class 'e' confusion matrix
    e_TP, e_FP, e_FN, e_TN = 0, 0, 0, 0
    pred = holdout_clf.predict(X_test)
    for i in range(len(X_test)):
        if pred[i] == 'e' and y_test[i] == 'e':
            e_TP += 1
        elif pred[i] == 'e' and y_test[i] == 'p':
            e_FP += 1
        elif pred[i] == 'p' and y_test[i] == 'e':
            e_FN += 1
        elif pred[i] == 'p' and y_test[i] == 'p':
            e_TN += 1
    confusion_mat1 = [[e_TP, e_FP], [e_FN, e_TN]]
    acc1 = (e_TP+e_TN)/len(X_test)
    recall1 = e_TP/(e_TP+e_FN)
    precision1 = e_TP/(e_TP+e_FP)
    
    # class 'p' confusion matrix
    p_TP, p_FP, p_FN, p_TN = e_TN, e_FN, e_FP, e_TP
    confusion_mat2 = [[f_TP, f_FP], [f_FN, f_TN]]
    acc2 = (f_TP+f_TN)/len(X_test)
    recall2 = f_TP/(f_TP+f_FN)
    precision2 = f_TP/(f_TP+f_FP)
    
    # avg. of 2 labels
    confusion_mat = (confusion_mat1+confusion_mat2)/2
    acc = (acc1+acc2)/2
    recall = (recall1+recall2)/2
    precision = (precision1+precision2)/2
    
    # Use score method to get accuracy of the model
    print('--------------------------------------------------------')
    holdout_score = model.score(X_test, y_test)
    print('Accuracy Score(holdout): ', holdout_score)
    print('--------------------------------------------------------')
    
    # Return relevant data for chart plotting
    return holdout_clf, holdout_score, confusion_mat, acc, recall, precision
```
